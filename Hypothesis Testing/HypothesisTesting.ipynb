# IMPORTS

```python
import math
import numpy as np
from scipy import stats
```

# Part A — Confidence Intervals (Print-Head Durability)

```python
# SAMPLE DATA
sample = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07,
                   1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n          = len(sample)           # sample size
alpha      = 0.01                  # significance level for 99% CI (1 - 0.99)
conf_level = 1 - alpha             # 0.99

sample_mean = sample.mean()
# ddof=1 : uses n-1 in denominator — gives the UNBIASED sample standard deviation
# WHY ddof=1 : dividing by n underestimates the true population std;
#              n-1 corrects for this (Bessel's correction)
sample_std  = sample.std(ddof=1)
se_sample   = sample_std / math.sqrt(n)   # Standard Error = s / sqrt(n)

print(f"n            = {n}")
print(f"Sample mean  = {sample_mean:.4f} million chars")
print(f"Sample std   = {sample_std:.4f} million chars")
print(f"Std Error    = {se_sample:.4f} million chars")
```

```python
# TASK a — 99% Confidence Interval using SAMPLE Standard Deviation (t-distribution)

df     = n - 1                              # degrees of freedom = 14

# t.ppf(1 - alpha/2, df) gives the t critical value for a TWO-SIDED 99% CI
# alpha/2 = 0.005 on each tail → total 1% outside the interval
t_crit = stats.t.ppf(1 - alpha/2, df)

ci_lower_t = sample_mean - t_crit * se_sample
ci_upper_t = sample_mean + t_crit * se_sample

print("=== Part A(a) — 99% CI using Sample Std Dev (t-distribution) ===")
print(f"Degrees of freedom        = {df}")
print(f"t critical value (99%)    = {t_crit:.4f}")
print(f"Margin of Error           = {t_crit * se_sample:.4f}")
print(f"99% CI = ({ci_lower_t:.4f},  {ci_upper_t:.4f})  million characters")
print()
print("Interpretation:")
print(f"  We are 99% confident the true mean durability lies between")
print(f"  {ci_lower_t:.4f} and {ci_upper_t:.4f} million characters.")
print()
print("Rationale:")
print("  n=15 (small sample) and population σ is unknown → use t-distribution.")
```

```python
# TASK b — 99% Confidence Interval using KNOWN Population Standard Deviation (z)
sigma_known = 0.2                            # given known population std dev
se_known    = sigma_known / math.sqrt(n)     # standard error using known σ

# norm.ppf(1 - alpha/2) gives the z critical value for a TWO-SIDED 99% CI
# For 99% CI: alpha=0.01, alpha/2=0.005 → z* ≈ 2.576
z_crit = stats.norm.ppf(1 - alpha/2)

ci_lower_z = sample_mean - z_crit * se_known
ci_upper_z = sample_mean + z_crit * se_known

print("=== Part A(b) — 99% CI using Known Population Std Dev (z-distribution) ===")
print(f"Known population sigma    = {sigma_known:.4f} million chars")
print(f"Standard Error            = {se_known:.4f}")
print(f"z critical value (99%)    = {z_crit:.4f}")
print(f"Margin of Error           = {z_crit * se_known:.4f}")
print(f"99% CI = ({ci_lower_z:.4f},  {ci_upper_z:.4f})  million characters")
print()
print("Interpretation:")
print(f"  We are 99% confident the true mean durability lies between")
print(f"  {ci_lower_z:.4f} and {ci_upper_z:.4f} million characters.")
print()
print("Note: When population σ is known → use z (normal) distribution.")
print("The z-interval is slightly narrower than the t-interval because")
print("there is no uncertainty from estimating σ.")
```

```python
# COMPARISON — t-interval vs z-interval
width_t = ci_upper_t - ci_lower_t
width_z = ci_upper_z - ci_lower_z

print("=== Comparison: t-interval vs z-interval ===")
print(f"{'Method':<35} {'Lower':>10} {'Upper':>10} {'Width':>10}")
print("-" * 65)
print(f"{'a) t-interval (σ unknown)':<35} {ci_lower_t:>10.4f} {ci_upper_t:>10.4f} {width_t:>10.4f}")
print(f"{'b) z-interval (σ = 0.2 known)':<35} {ci_lower_z:>10.4f} {ci_upper_z:>10.4f} {width_z:>10.4f}")
print()
print(f"The t-interval is {width_t - width_z:.4f} million chars wider than the z-interval.")
print("This is expected — the t-distribution has heavier tails to account")
print("for the extra uncertainty of estimating σ from a small sample.")
```

# Part B — Hypothesis Testing (Bombay Hospitality Ltd.)

```python
# GIVEN DATA — Bombay Hospitality
# Cost model: W = 1000 + 5X
# X = number of units produced per week
# X ~ Normal(mean=600, std=25)
# Sample: n=25 restaurants, sample mean weekly cost = Rs 3050

X_mean    = 600          # population mean of units produced
sigma_X   = 25.0         # population std dev of units produced
n_sample  = 25           # number of restaurants sampled
xbar      = 3050.0       # observed sample mean weekly cost
W_model_mean = 1000 + 5 * X_mean
sigma_W = 5 * sigma_X

print("=== Part B — Given Data ===")
print(f"Cost model              : W = 1000 + 5X")
print(f"X mean (μ_X)            = {X_mean}")
print(f"X std dev (σ_X)         = {sigma_X}")
print(f"Theoretical W mean (μ_W)= 1000 + 5×{X_mean} = {W_model_mean:.2f}")
print(f"Population σ_W          = 5 × {sigma_X} = {sigma_W:.2f}")
print(f"Sample size n           = {n_sample}")
print(f"Sample mean (x̄)         = {xbar:.2f}")
```

```python
# STEP 1 — STATE THE HYPOTHESES
print("=== STEP 1 — Hypotheses ===")
print()
print("H₀ (Null Hypothesis)        : μ = 4000")
print("   The weekly cost follows the model W = 1000 + 5X")
print("   and the true mean is Rs 4000 when X̄ = 600.")
print()
print("H₁ (Alternative Hypothesis) : μ > 4000  (one-sided upper-tail test)")
print("   The restaurant owners claim actual weekly costs are")
print("   HIGHER than what the model predicts.")
print()
print("Significance level (α) = 0.05")
```

```python
# STEP 2 — CALCULATE THE TEST STATISTIC (z)
alpha_ht = 0.05                             # significance level
se_w     = sigma_W / math.sqrt(n_sample)   # standard error of the mean
z_stat   = (xbar - W_model_mean) / se_w    # test statistic

print("=== STEP 2 — Test Statistic ===")
print()
print(f"Standard Error  SE = σ_W / √n = {sigma_W} / √{n_sample} = {se_w:.4f}")
print(f"z statistic        = (x̄ - μ₀) / SE")
print(f"                   = ({xbar} - {W_model_mean}) / {se_w:.4f}")
print(f"                   = {xbar - W_model_mean:.2f} / {se_w:.4f}")
print(f"                   = {z_stat:.4f}")
print()
print("Interpretation:")
print(f"  The sample mean ({xbar}) is {abs(z_stat):.2f} standard errors BELOW")
print(f"  the hypothesised mean ({W_model_mean}).")
```

```python
# STEP 3 — DETERMINE THE PROBABILITY (p-value) AND CRITICAL VALUE
z_crit_one = stats.norm.ppf(1 - alpha_ht)

# One-sided p-value: P(Z > z_stat) for upper-tail test H₁: μ > μ₀
# norm.cdf(z_stat) gives P(Z ≤ z_stat), so 1 - cdf gives P(Z > z_stat)
p_one_sided = 1 - stats.norm.cdf(z_stat)

# Two-sided p-value (shown for completeness)
p_two_sided = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("=== STEP 3 — Critical Value and p-value ===")
print()
print(f"Significance level α         = {alpha_ht}")
print(f"One-sided critical z (α=0.05)= {z_crit_one:.4f}")
print(f"Calculated z statistic       = {z_stat:.4f}")
print()
print(f"One-sided p-value            = {p_one_sided:.4e}")
print(f"Two-sided p-value            = {p_two_sided:.4e}  (shown for reference)")
print()
print("Interpretation of p-value:")
if p_one_sided < alpha_ht:
    print(f"  p-value ({p_one_sided:.4f}) < α ({alpha_ht}) → result IS statistically significant.")
else:
    print(f"  p-value ({p_one_sided:.4f}) > α ({alpha_ht}) → result is NOT statistically significant.")
```

```python
# STEP 4 — MAKE A DECISION
print('=== STEP 4 — Decision ===')
print()
print(f'Decision rule : Reject H0 if z_stat > {z_crit_one:.4f}')
print(f'Calculated z  = {z_stat:.4f}')
print(f'Critical z    = {z_crit_one:.4f}')
print()

if z_stat > z_crit_one:
    decision = 'REJECT H0'
    detail   = 'The test statistic falls in the rejection region.'
else:
    decision = 'FAIL TO REJECT H0'
    detail   = 'z_stat is NOT greater than z_crit — does not fall in the rejection region.'

print(f'Decision: {decision}')
print(f'  {detail}')
```

```python
# STEP 5 — CONCLUSION
# Interpret the statistical decision in plain business language.
# --> Statistical results must be translated into actionable insightscfor non-technical stakeholders (e.g. management, restaurant owners).
print("=== STEP 5 — Conclusion ===")
print()
print("Statistical Conclusion:")
print("  At the 5% significance level, there is NO statistically significant")
print("  evidence to support the restaurant owners' claim that weekly")
print("  operating costs are higher than the model W = 1000 + 5X predicts.")
print()
print("Business Interpretation:")
print(f"  The sample mean weekly cost (Rs {xbar:.0f}) is actually LOWER than")
print(f"  the theoretical model mean (Rs {W_model_mean:.0f}). The observed data")
print(f"  contradicts the owners' claim — costs appear to be BELOW, not above,")
print("  the model prediction.")
print()
print("  The extremely negative z-statistic (z = -38.00) indicates the sample")
print("  mean is 38 standard errors BELOW the model mean. This strongly")
print("  suggests the model overestimates costs, not underestimates them.")
print()
print("Recommendation:")
print("  Management should investigate whether:")
print("  1. The cost model needs to be revised downward.")
print("  2. The sample is representative (possible sampling bias).")
print("  3. The sample mean (3050) includes all cost components.")

# Full summary table
print()
print("=" * 60)
print("COMPLETE SUMMARY")
print("=" * 60)
print(f"  H₀            : μ = {W_model_mean:.0f}")
print(f"  H₁            : μ > {W_model_mean:.0f}  (one-sided)")
print(f"  Sample mean   : Rs {xbar:.0f}")
print(f"  σ_W           : {sigma_W:.0f}")
print(f"  n             : {n_sample}")
print(f"  SE            : {se_w:.4f}")
print(f"  z statistic   : {z_stat:.4f}")
print(f"  z critical    : {z_crit_one:.4f}")
print(f"  p-value       : {p_one_sided:.4e}")
print(f"  α             : {alpha_ht}")
print(f"  Decision      : {'Reject H₀' if z_stat > z_crit_one else 'Fail to Reject H₀'}")
print("=" * 60)
```

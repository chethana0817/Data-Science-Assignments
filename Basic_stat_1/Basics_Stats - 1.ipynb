## Data Visulization

### For data visulization there is a python libraries used Matplotlib,Seaborn,.

### Matplotlib ---> Matplotlib is used to convert the dataset in table form to visulized form.It is used to create graphs and chats for better understanding of gievn dataset. 
### Seaborn ---> its a python built in library on matplotlib , this is used to create colorfull charts and graphs.
### Along with this visualized libraries we should also "import warinings" which is used to keep output clean without displaying warning messages,'warnings.filterwarnings("ignore")' this tells to python if there is any warnings ,it should be ignored(like hide all warnings and keep output clean).

```python
# first let me import matplotlib library if not there use 'pip install matplotlib' to install the library
import matplotlib.pyplot as plt
# pyplot is a module in matplotlib which helps to create graphs and charts easily, 'as' --> alias name for matplotlib.pyplot as 'plt'
import seaborn as sns  # sns is alias name for seaborn
import warnings
warnings.filterwarnings("ignore")
```

### Histogram ---> a histogram is a graphical representation of the distribution of numberical data grouping values into ranges.
### Working ---> x-axis represents ranges of values. y-axis represents the frequency(count of values) in each range.
### each bar shows how many data points fall within that range.
### bins ---> it shows how frequently data values occur within specific range.

```python
df.hist(figsize=(15, 10), bins=20) 
# df.hist() -->automatically detects numeric features and creates one histogram per column.
# fisize=(15,10) -->sets the width and height of the overall figure in inches, making the plots larger and readable.
# bins =20 ---> divides the data into 20 equal intervals,controlling how detailed the distribution appears.
plt.suptitle("Histogram of All Features", fontsize=16)
# suptitle ---> adds a main title to the entire figure.
# fontsize=16 ---> increases the font size for better visibility.
plt.ylabel('Frequency')
# ylabel ---> sets the y-axis label for all histograms.
# 'Frequency' ---> represents how many data values fall into each bin.
plt.tight_layout()
# tight_layout() ---> automatically adjusts spacing between subplots(avoids overlapplingof titles)
plt.show()
# show() ---> displays the final output on the screen.
```

### box plot ---> also called as box and whisker plot its a statistical graph used to show the distribution of numberical data based on five keys values.
### 'minimum' -->smallest value(exluding outliers),
### 'Q1' ---> 25% of data lies below this Q1, 'Q2'---> middle value of the dataset, 'Q3' ---> 75% of data lies below this Q3 
### 'maximum' ---> largest value(excluding outliers) 

```python
plt.figure(figsize=(15, 6)) 
## figure() ---> creates new figure for the plot.
## figsize(15,6) ---> sets the width to 15 inches and height to 6 inches,making the plot wide and readable when many features are present.
sns.boxplot(data=df)
## boxplot() ---> this boxplot function is taken from seaborn library. which is used to draw box plot for all numerical colimns in the dataframe.
## data=df ---> tells the seaborn to use the dataset df for plotting
## each box represent one numerical feature and shows its distribution.
plt.xticks(rotation=90)
## this x-axis used to rotate all the lables below the boc plot to 90 degrees.
## its used to prevent overlapping of many columns. 
plt.title("Boxplot of All Features")
## title() ---> adds a title to the plot.
plt.show()
```

### Bar chart ---> its a graphical representation of categorical data,that shows the frequency or count of each category using rectangular bars.
### x-axis---> shows the categories(data,day,city,brand,..etc...)
### y-axis --->frequency or count of each category
### bars ---> this reprsents , greater the height of bar occurrences of values is more. 

```python
calc_stat1 = df.select_dtypes(include=['object']).columns
##created new variable 'calc_stat1' to store categorical columns of datatype'object' only.
```

```python
calc_stat1 ## shows only categorical columns selecting columns based on datatype 'object'.
```

```python
for col in calc_stat1: 
    plt.figure(figsize=(7,5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show() 
## for loop is used because in categorical columns the value_counts will differ for different columns.
## value_counts() ---> works on one column at a time.
## so, we must loop through each columns to plot the bar separately.
```

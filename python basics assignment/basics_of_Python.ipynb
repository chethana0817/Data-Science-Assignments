# Exercise 1: Prime Numbers

```python
num=int(input('enter a number: ')) # used variable num and int datatype while user enter the input number 
if num <=1: # checking if the user entered num is <=1 ,if less not a prime number
	print('not a prime number') # display to user output and exit 
else: # if user num is greater than 1 this else part is executed
	is_prime =True # initially assuming its prime 
	for i in range(2,int(num **0.5)+1): #then for loops is used from 2 to the square root of the num
		if num % i==0: # checks if the number is divisible by i
			is_prime = False # then sets is_prime to False if a divisor is found
			break # if not prime this stops the loop
	if  is_prime: # then if its prime 
		print('Prime number') # this print stmt is executed
	else:  # if not prime ,else part is executed
		print('not a prime number')
```

# Exercise 2: Product of Random Numbers

```python
# let me first import random module
import random
# let me take 2 var(num1 and num2 to generate random variables between 1 and 10)
num1 =random.randint(1,10)
num2 =random.randint(1,10)
# let me take var(user_ans) so that to store the user product value
user_ans =int(input(f'what is the product of {num1} and {num2}? '))
# check the user product value is correct r not
if user_ans == num1*num2:
    print('correct product value!')
else: #if not correct
    print('incorrect')
    #  print the actual product value of the random values
    print(f'the correct answer is {num1 * num2}') 
```

# Exercise 4:Problem Statement (Question): Hospital Billing System

```python
# hospital charges
room_charge_per_day =2000
doctor_fee =1500
lab_test_charge=300
#patient details
days_stayed=4
lab_tests=3
medicine_charges=2400
#calculate individual costs
room_charges=room_charge_per_day*days_stayed
lab_charges=lab_test_charge*lab_tests
#total bill before discount

total_bill=room_charges+doctor_fee+lab_charges+medicine_charges
#apply discount if total exceeds 10,000
if total_bill>10000:
    discount =total_bill*0.10
else:
    discount=0
    
# hospital final bill
final_bill=total_bill-discount
#display bill details
print("Hospital Bill Details")
print("Room Charges:",room_charges)
print("Doctor Fee:",doctor_fee)
print("Lab charges:",lab_charges)
print("Medicine charges:",medicine_charges)
print("Discount:",discount)
print("Final Bill Amount:",final_bill)
```

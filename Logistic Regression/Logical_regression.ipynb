# 1.Import Libraries

```python
# pandas library is used for data manipulation and analysis
# It helps in loading datasets, cleaning data, and performing operations on data tables (DataFrames)
import pandas as pd

# numpy is used for numerical computations and mathematical operations on arrays
# It is commonly used for handling large numerical datasets efficiently
import numpy as np

# matplotlib is a visualization library used to create graphs and charts
# pyplot module provides functions to create plots like line plots, histograms, etc.
import matplotlib.pyplot as plt

# seaborn is built on top of matplotlib and is used for advanced statistical data visualization
# It helps create more attractive plots such as heatmaps, pairplots, and boxplots
import seaborn as sns

# train_test_split is used to divide the dataset into training data and testing data
# Training data is used to train the model, while testing data is used to evaluate model performance
from sklearn.model_selection import train_test_split

# StandardScaler is used to standardize numerical features
# It transforms data so that the mean becomes 0 and the standard deviation becomes 1
# This helps machine learning models perform better when features have different scales
from sklearn.preprocessing import StandardScaler

# LogisticRegression is the machine learning algorithm used for binary classification
# It predicts probabilities and classifies data into two categories (e.g., diabetes or no diabetes)
from sklearn.linear_model import LogisticRegression

# accuracy_score calculates the percentage of correct predictions made by the model
# precision_score measures how many predicted positive cases are actually correct
# recall_score measures how many actual positive cases were correctly identified
# f1_score is the harmonic mean of precision and recall
# roc_auc_score measures the model's ability to distinguish between classes
# roc_curve is used to plot the Receiver Operating Characteristic curve
# confusion_matrix shows the number of correct and incorrect predictions
# classification_report provides a detailed report including precision, recall, f1-score, and accuracy
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
```

# 2.Load Dataset

```python
# pd.read_csv() is a pandas function used to load a dataset from a CSV (Comma Separated Values) file
# It reads the file "diabetes.csv" and stores it in a DataFrame called 'data'
# A DataFrame is a table-like structure with rows and columns used for data analysis
data = pd.read_csv("diabetes.csv")

# data.head() displays the first 5 rows of the dataset
# It is used to quickly inspect the dataset structure, column names, and sample values
# This helps verify that the dataset was loaded correctly
data.head()
```

# 6. Data Preprocessing

```python
# Now Replace zeros with NaN for specific columns where 0 is not realistic
# In this dataset, some features like Glucose, BloodPressure, BMI, etc.
# cannot realistically be zero. A value of 0 often indicates missing data.
# Therefore, replace these zeros with NaN (Not a Number) to treat them as missing values.

cols_to_replace = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# data[cols_to_replace] selects only the specified columns
# replace(0, np.nan) replaces all zero values with NaN in those columns
data[cols_to_replace] = data[cols_to_replace].replace(0, np.nan)
```

# Fill missing values with median

```python
# fillna() replaces missing values (NaN) with a specified value
# data.median() calculates the median for each column
# The median is used instead of the mean because it is less affected by outliers
# inplace=True updates the original dataset directly
data.fillna(data.median(), inplace=True)
```

# Split features and target

```python
# data.drop('Outcome', axis=1) removes the target column from the dataset
# The remaining columns become the independent variables (features)
X = data.drop('Outcome', axis=1)

# The Outcome column is the dependent variable (target)
# It indicates whether a person has diabetes (1) or not (0)
y = data['Outcome']
```

# Standardize features

```python
# StandardScaler is used to standardize numerical features
# It transforms data so that the mean becomes 0 and the standard deviation becomes 1
# This is important because machine learning models perform better
# when features are on a similar scale

scaler = StandardScaler()

# fit_transform() first calculates the scaling parameters (mean and std)
# and then applies the transformation to the dataset
X_scaled = scaler.fit_transform(X)
```

# Split into train and test

```python
# train_test_split divides the dataset into training and testing sets
# X_scaled = feature variables
# y = target variable
# test_size=0.2 means 20% of data will be used for testing
# random_state=42 ensures reproducibility of results

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

# 7.Build and Train Logistic Regression Model

```python
# LogisticRegression() creates a logistic regression model object
# Logistic regression is a classification algorithm used to predict binary outcomes
# In this dataset, it predicts whether a person has diabetes (1) or not (0)
model = LogisticRegression()

# model.fit() trains the logistic regression model using the training data
# X_train contains the feature variables (input data)
# y_train contains the target variable (actual diabetes labels)
# The model learns the relationship between features and the target variable
model.fit(X_train, y_train)

# model.predict() predicts the class labels for the test dataset
# It outputs either 0 (No Diabetes) or 1 (Diabetes)
# These predicted values are used to evaluate model performance
y_pred = model.predict(X_test)


# model.predict_proba() returns the probability of each class
# [:,1] selects the probability of the positive class (Diabetes = 1)
# These probabilities are used for evaluation metrics like ROC-AUC and ROC curve
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

#### In this step, I built a logistic regression model and trained it using the training dataset. Then predictions and probability scores on the test dataset to evaluate the model using different classification metrics

# 8.Model Evaluation

```python
# accuracy_score calculates the proportion of correct predictions made by the model
# It compares the actual values (y_test) with predicted values (y_pred)
# Accuracy = (Correct Predictions) / (Total Predictions)
print("Accuracy:", accuracy_score(y_test, y_pred))


# precision_score measures how many predicted positive cases are actually correct
# Precision = TP / (TP + FP)
# It is important when the cost of false positives is high
print("Precision:", precision_score(y_test, y_pred))


# recall_score measures how many actual positive cases were correctly identified
# Recall = TP / (TP + FN)
# It is important when missing positive cases is costly (e.g., disease detection)
print("Recall:", recall_score(y_test, y_pred))


# f1_score is the harmonic mean of precision and recall
# It balances both precision and recall and is useful when dataset classes are imbalanced
print("F1 Score:", f1_score(y_test, y_pred))


# roc_auc_score measures the model's ability to distinguish between classes
# It uses probability predictions instead of class labels
# A value close to 1 indicates a strong model, while 0.5 indicates a random model
print("ROC AUC:", roc_auc_score(y_test, y_pred_proba))


# Confusion Matrix

# confusion_matrix creates a matrix showing the number of correct and incorrect predictions
# It contains four values:
# True Positive (TP), True Negative (TN), False Positive (FP), False Negative (FN)
cm = confusion_matrix(y_test, y_pred)
```

### I evaluated the logistic regression model using metrics such as accuracy, precision, recall, F1 score, and ROC-AUC. I also used a confusion matrix to visualize the model's classification performance

```python
# sns.heatmap() visualizes the confusion matrix using a heatmap
# annot=True displays numbers inside the cells
# fmt='d' formats the numbers as integers
# cmap='Blues' sets the color style
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')


# plt.title() adds a title to the plot
plt.title("Confusion Matrix")


# plt.xlabel() labels the x-axis (predicted values)
plt.xlabel("Predicted")


# plt.ylabel() labels the y-axis (actual values)
plt.ylabel("Actual")


# plt.show() displays the confusion matrix plot
plt.show()
```

#### The logistic regression model was evaluated using several performance metrics. Accuracy measures the overall correctness of the model, while precision and recall evaluate the performance on positive class predictions. The F1 score provides a balance between precision and recall. The ROC-AUC score indicates how well the model distinguishes between diabetic and non-diabetic cases. The confusion matrix visually represents the classification results

```python
import pickle
# pickle is a Python library used to serialize (save) Python objects such as machine learning models

with open("diabetes_model.pkl", "wb") as file:
    # open() creates a new file called "diabetes_model.pkl"
    # "wb" means write binary mode, which is required to store machine learning models

    pickle.dump(model, file)
    # pickle.dump() saves the trained logistic regression model into the file
    # This allows us to reuse the model later without retraining it
```

### Pickle is used to serialize machine learning models. After training the model,we save it as a .pkl file so it can be loaded later in applications like Streamlit without retraining.

# 9.ROC Curve

```python
# roc_curve() computes the False Positive Rate (FPR) and True Positive Rate (TPR)
# at different classification thresholds using the actual labels (y_test)
# and predicted probabilities (y_pred_proba)
# "_" stores the threshold values which are not needed here
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)


# plt.figure() creates a new figure for plotting
# figsize=(6,5) sets the size of the ROC curve plot
plt.figure(figsize=(6,5))


# plt.plot() draws the ROC curve using FPR (x-axis) and TPR (y-axis)
# The label shows the ROC-AUC score which indicates model performance
# roc_auc_score() calculates the Area Under the Curve value
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc_score(y_test, y_pred_proba):.2f})")


# This line plots a diagonal dashed line representing a random classifier
# If the model performs no better than random guessing, it will follow this line
plt.plot([0,1], [0,1], 'k--')


# plt.xlabel() labels the x-axis as False Positive Rate
# FPR indicates the proportion of negative cases incorrectly classified as positive
plt.xlabel("False Positive Rate")


# plt.ylabel() labels the y-axis as True Positive Rate
# TPR indicates the proportion of actual positive cases correctly identified
plt.ylabel("True Positive Rate")


# plt.title() adds a title to the ROC curve plot
plt.title("ROC Curve")


# plt.legend() displays the legend showing the ROC curve label and AUC score
plt.legend()


# plt.show() displays the ROC curve plot
plt.show()
```

### The ROC curve was plotted to evaluate the classification performance of the logistic regression model. The ROC curve shows the trade-off between the true positive rate and false positive rate at different classification thresholds. The area under the curve (AUC) indicates the model's ability to distinguish between diabetic and non-diabetic patients

# 10.Interpretation - Coefficients

```python
# pd.DataFrame() creates a new DataFrame to display the model coefficients
# "Feature": X.columns stores the names of the input features used in the model
# "Coefficient": model.coef_[0] retrieves the coefficient values learned by the logistic regression model
# Each coefficient represents how strongly a feature influences the prediction
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_[0]
})


# sort_values() sorts the features based on coefficient values
# by="Coefficient" sorts according to coefficient magnitude
# ascending=False sorts from highest to lowest importance
# This helps identify which features have the strongest influence on the prediction
coefficients = coefficients.sort_values(by="Coefficient", ascending=False)


# model.intercept_ returns the intercept (bias term) of the logistic regression model
# The intercept represents the baseline value when all feature values are zero
print("Intercept:", model.intercept_[0])


# Print heading for feature importance table
print("\nFeature Importance:")


# Print the DataFrame containing features and their coefficients
# This helps interpret which features contribute most to predicting diabetes
print(coefficients)
```

### The coefficients of the logistic regression model were analyzed to understand the influence of each feature on diabetes prediction. Features with larger positive coefficients increase the probability of diabetes, while negative coefficients reduce the probability. This helps identify the most important predictors in the dataset.

# 11.Summary Report

```python
# Print a heading to indicate that the following output is the classification report
print("Classification Report:")


# classification_report() generates a detailed performance report for the model
# It compares the actual values (y_test) with the predicted values (y_pred)

# The report includes important evaluation metrics such as:
# Precision  → How many predicted positive cases are actually correct
# Recall     → How many actual positive cases were correctly identified
# F1-score   → Balance between precision and recall
# Support    → Number of actual samples in each class

# This summary helps evaluate how well the logistic regression model performs
print(classification_report(y_test, y_pred))
```

### The classification report provides detailed performance metrics for the logistic regression model. It includes precision, recall, and F1-score for each class. These metrics help evaluate how effectively the model distinguishes between diabetic and non-diabetic patients.

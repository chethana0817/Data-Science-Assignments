```python
df.dtypes ## checking datatype of each columns
```

```python
for colm in df.columns:
    if df[colm].dtype =='object':
        try:
            df[colm]=df[colm].astype(float)
        except:
            pass
## loops through all columns and checks if any columns datatype is 'object', if found it will convert to float 
## if not found 'pass' is used to skip the columns safely.
```

```python
sns.boxplot(data=df) ## draws boxplots for all columns(features) and used to find outliers in dataframe.
plt.xticks(rotation=90) ## roatates the lables of x-axis to 90 degree for easier readable without any overlapping.
plt.show()
## all small empty circles are considered as outliers.
```

```python
## quantile used to know the data spread ,used to detect outliers.
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | 
                      (df > (Q3 + 1.5 * IQR))).any(axis=1)]## used here to remove extreme outliers values
## to imporve model acuracy and visualizaton clarity.
```

```python
df.shape, df_no_outliers.shape 
## used here to compare between orginal dataset size and dataset without outliers.
```

```python
sns.boxplot(data=df) ##before removing  of extreme values(outliers) in dataset viewed using boxplot
plt.show()

sns.boxplot(data=df_no_outliers) ##after removing  of extreme values(outliers) in dataset viewed using boxplot
plt.show() 
```

## 3.Data Visualization:

```python
df.hist(figsize=(15,12)) ## histogram showing how numberical data is distributed  
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df) ## for detecting outliers.
plt.title("Boxplots for Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15, 10)) 
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='cividis') ## heatmap used to find the relationship between the variables(features) in dataset
plt.title("Correlation Heatmap")
plt.show()
## correlation heatmap measures how strongly 2 variables are related to each other.
## +1 value ---> perfect postive relationship between 2 variables.
## 0 value ---> no relationship between 2 variables.
## -1 value ---> perfect negative relationship between 2 variables.
```

```python
sns.pairplot(df.sample(min(200, len(df))), diag_kind='kde',plot_kws={'color': 'teal'},
    diag_kws={'color': 'darkred'}) 
plt.suptitle("Pairplot of Selected Variables" , fontsize=26, y=1.02 , fontweight= 'bold') ## y=1.02 used to move slightly upwards, it will not overlap with plots.
plt.show()
## pairplot is used between numerical variables which shows relationship between every pair of features.
## sample will take random rows in the it will select min 200 rows to keep pairplot fast and error-free.
# diag_kind='kde' used to Display 'Kernel Density Estimation' (KDE) plots on the diagonal which is used for showing smooth distribution curves.
# plot_kws={'color': 'teal'} used to set the color of the none-diagonal scatter plots to teal to make it visual clarity and make plot understand batter.
# diag_kws={'color': 'darkred'} used to  set the color of the diagonal KDE curves to dark red , making distribution patterns more visible and easier to understand.
```

#### Violin plot is used to see the data distribution,density, and spread of data in each single plots
#### wider part shows ---> more data points spreaded 
#### Narrow part shows---> fewer data points spreaded.

```python
plt.figure(figsize=(14, 8)) 
sele_feat = df.columns[:6] ## select columns [0:6] stores it in 'sele_feat', only 6 columns is used to limit analysis, plotting all columns would be messy.
for feature in sele_feat: ## from features selecting only 'sele_feat' columns
    sns.violinplot(x=feature, data=df ,color='skyblue') ## now violin plot is created for current features.,x=features values and data is taken from dataframe.
    plt.title(f"Violin Plot for {feature}")
    plt.show()
```

```python
corr = df.corr() ## calculates the correlation for all numerical columns.
high_corr = corr[(corr > 0.7) & (corr < 1.0)] ## (filtering step) if correlation is  greater than 0.7 it shows strong correlation, if less than 1.0 its excludes self -correlation. 
print("\nHighly Correlated Features (corr > 0.7):") 
print(high_corr.dropna(how='all').dropna(axis=1, how='all')) ## removes rows where all values are NaN for clear output.
## dropna(axis=1) removes columns where all values are NaN, keeps only columns that actually have strong correlations.
```

#### if no features are highly correlated then there is no multicollinearity and features are mostly independent.
#### multicollinearity means when two or more independent variables are highly correlated with each other(they carry almost the same information)
#### If multicollinearity is there then 
#### 1. confuses the model (Model can't decide which feature is more important)
#### 2. reduces feature importances and becomes unreliable and so on....
### so to avoid such multicollinearity related problems , correlations, heatmap is used to detect correlation features and drop one feature from each correlated pairs and futher analysis is performed. to make to model more efficient.
